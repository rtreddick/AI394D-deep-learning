```python
import matplotlib.pyplot as plt
import torch

def f(x: torch.Tensor) -> torch.Tensor:
    return 3 * x ** 2 + torch.sin(5 * x)


def grad(x: torch.Tensor) -> torch.Tensor:
    return 6 * x + 5 * torch.cos(5 * x)


plt.plot(torch.arange(-1, 1, 0.01), f(torch.arange(-1, 1, 0.01)))
plt.plot(torch.arange(-1, 1, 0.01), grad(torch.arange(-1, 1, 0.01)))
```

```python
mu = 0
sigma = 3
x = mu + sigma * torch.randn(1)
eta = 0.02

plt.plot(torch.arange(-2, 2, 0.01), f(torch.arange(-2, 2, 0.01)))

all_x = []
for _ in range(100):
    x = x - eta * grad(x)
    all_x.append(x)

plt.plot(all_x, f(torch.tensor(all_x)), 'ro')
```
